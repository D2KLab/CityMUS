```python
import csv
import json
with open('final_paths.csv','r') as input_fp:
    reader=csv.reader(input_fp,)
    # skip header
    next(reader)
    rows = [ [col for col in row] for row in reader]

with open('artists_tracks.json','r') as input_fp:    
    artists_tracks = json.load(input_fp)    
    
```

```python
import pandas as pd
final_paths = pd.read_csv('final_paths.csv')
```

```python
artists = {artists_tracks[key]['dbpedia_uri']:artists_tracks[key] for key in artists_tracks.keys()}
```

```python
poi_artist_path = dict()
for row in rows:
    poi_artist_path[row[12]] = dict()
```

```python
for row in rows:
    if len(poi_artist_path[row[12]]) < 5:
        try:
            if row[0] in artists:
                poi_artist_path[row[12]][row[0]]
        except KeyError:
            path_tracks = dict()
            path = row[:13]
            tracks = artists[row[0]]['tracks']
            path_tracks['path'] = path
            path_tracks['tracks'] = tracks
            print(len(tracks))
            poi_artist_path[row[12]][row[0]] = path_tracks
```

```python
with open('poi_artist_path.json','w') as output_fp:
    json.dump(poi_artist_path,output_fp)
```

```python
import numpy as np
import pickle
selected_paths = []
for key_1 in poi_artist_path.keys():
    for key_2 in poi_artist_path[key_1].keys():
        selected_paths.append(poi_artist_path[key_1][key_2]['path'])
pickle.dump( selected_paths, open( "selected_paths.p", "wb" ) )
```
